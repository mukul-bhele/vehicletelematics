# Extracting Accelerometer & Magnetometer Data

**accData** – It refers to Accelerometer and Magnetometer sensor data. The data is collected from the OBD device, values are in terms of G-force. The data is across X, Y, Z axis where X-axis is horizontal, Y- axis is vertical and Z-axis is the direction of movement of the car. 
The data is provided in raw format. To extract values, we have use certain manipulations: 

```python
import pandas as pd
import numpy as np
import math

# Convert Accelerometer reading to g units, assuming 8-bit data with a range of +-2g
def convert_acc(x):
    x = int(x, 16)
    if x > 127:
        x = x - 256
    return np.float64(x * 2 / 128)

# Convert Magnetometer reading to uT units, assuming 16-bit data with a range of +-1200uT
def convert_mag(x):
    x = int(x, 16)
    if x > 32767:
        x = x - 65536
    return np.float64(x * 0.0366)

# Convert raw ecompass data to accelerometer and magnetometer data
def convert_ecompass(raw_data):
    if len(raw_data) >= 18:
        acc_data = {
            'x': convert_acc(raw_data[0:2]),
            'y': convert_acc(raw_data[2:4]),
            'z': convert_acc(raw_data[4:6])
        }
        mag_data = {
            'mx': convert_mag(raw_data[6:10]),
            'my': convert_mag(raw_data[10:14]),
            'mz': convert_mag(raw_data[14:18])
        }
        return acc_data, mag_data
    else:
        return None, None

# Function to calculate rotation matrix using complementary filter
def calculate_rotation_matrix(accel_data, mag_data, alpha=0.98):
    acc_norm = np.linalg.norm([accel_data['x'], accel_data['y'], accel_data['z']])
    if acc_norm == 0:
        return np.identity(3)
    
    accel_data_normalized = {
        'x': accel_data['x'] / acc_norm,
        'y': accel_data['y'] / acc_norm,
        'z': accel_data['z'] / acc_norm
    }

    cross_product = np.cross(list(accel_data_normalized.values()), list(mag_data.values()))
    cross_norm = np.linalg.norm(cross_product)
    
    if cross_norm == 0:
        return np.identity(3)

    accel_data_normalized = np.array(list(accel_data_normalized.values()))
    cross_product = cross_product / cross_norm

    rotation_matrix = np.array([
        cross_product,
        accel_data_normalized,
        np.cross(cross_product, accel_data_normalized)
    ]).T

    return alpha * rotation_matrix + (1 - alpha) * np.identity(3)

# Apply the conversion and rotation matrix calculation to each row of 'accData' column
for index, row in df.iterrows():
    acc_data = row['accData']
    
    if acc_data == '0':
        df.at[index, 'x'] = 0
        df.at[index, 'y'] = 0
        df.at[index, 'z'] = 0
        df.at[index, 'mx'] = 0
        df.at[index, 'my'] = 0
        df.at[index, 'mz'] = 0
        continue  # Skip rotation matrix calculation for accData = 0
    
    accel_data, mag_data = convert_ecompass(acc_data)
    
    if accel_data and mag_data:
        rotation_matrix = calculate_rotation_matrix(accel_data, mag_data)
        
        # Calculate x, y, z and mx, my, mz from the rotation matrix
        x, y, z = rotation_matrix.dot(np.array([1, 0, 0]))
        mx, my, mz = rotation_matrix.dot(np.array([0, 0, 1]))
        
        # Update respective columns in the DataFrame
        df.at[index, 'x'] = x
        df.at[index, 'y'] = y
        df.at[index, 'z'] = z
        df.at[index, 'mx'] = mx
        df.at[index, 'my'] = my
        df.at[index, 'mz'] = mz
        
        # Store the rotation matrix back into 'accData' column
        df.at[index, 'accData'] = rotation_matrix.tolist()
    else:
        df.at[index, 'accData'] = '0'
# Print the updated DataFrame
print(df)
```

This code defines two functions, `convert_acc(x)` and `convert_mag(x)`, which are used to convert raw sensor readings into more meaningful units. These functions are commonly used in sensor data processing, especially for applications like orientation estimation and motion tracking. Here's an explanation of each function:

**1. `convert_acc(x)` Function:**
   - This function is designed to convert accelerometer readings to units of gravitational force, typically measured in "g" (gravity). It assumes that the accelerometer data is in 8-bit format with a range of +/-2g.

   - Parameters:
     - `x`: This parameter represents the raw accelerometer reading in hexadecimal format.

   - Function Logic:
     - It first converts the hexadecimal input `x` into an integer using `int(x, 16)`. This step is necessary to convert the hexadecimal data into a numerical format.

     - Next, it checks if the integer value `x` is greater than 127. If `x` is greater than 127, it means that the value is negative and should be adjusted accordingly. This adjustment is done by subtracting 256 from `x`.

     - Finally, the function scales the adjusted `x` value to represent gravitational force. It multiplies the value by `2 / 128` to obtain the acceleration in units of g.

   - Return Value:
     - The function returns the converted accelerometer reading as a NumPy float.

**2. `convert_mag(x)` Function:**
   - This function is intended to convert magnetometer readings to units of magnetic field strength, typically measured in microteslas (uT). It assumes that the magnetometer data is in 16-bit format with a range of +/-1200uT.

   - Parameters:
     - `x`: This parameter represents the raw magnetometer reading in hexadecimal format.

   - Function Logic:
     - Similar to the `convert_acc` function, it first converts the hexadecimal input `x` into an integer.

     - It then checks if the integer value `x` is greater than 32767, indicating a negative value. If this condition is met, it adjusts `x` by subtracting 65536 to make it negative.

     - Finally, the function scales the adjusted `x` value to represent magnetic field strength in uT. It multiplies the value by `0.0366` to achieve this conversion.

   - Return Value:
     - The function returns the converted magnetometer reading as a NumPy float.

In summary, these functions are used to handle raw sensor data, correct for potential negative values, and convert the data into physically meaningful units (g for acceleration and uT for magnetic field strength) for further analysis or use in various applications.

**3. `calculate_rotation_matrix` Function:**

   - The main function calculate_rotation_matrix computes the rotation matrix using complementary filtering. It first normalizes accelerometer data to obtain the device's orientation in 3D space. Then, it calculates the cross product between the normalized accelerometer and magnetometer data, which provides the orientation of the device in relation to the Earth's magnetic field. The final rotation matrix is a combination of these values, weighted by an alpha parameter.

**4. Iterating Through Datarame :** 


   - The code iterates through each row in the DataFrame ('df'). For each row, it retrieves the accelerometer and magnetometer data. If the accelerometer data is valid (not '0'), it calculates the rotation matrix using the calculate_rotation_matrix function. This rotation matrix is then used to derive orientation values (x, y, z, mx, my, mz). These orientation values are updated in the DataFrame.
